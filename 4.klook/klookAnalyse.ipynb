```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif'] = ['SimHei']
from datetime import datetime
```

```python
df = pd.read_csv('./data/kelu.csv')
df.info()
```

```python
df.describe()
```

```python
df['time']=pd.to_datetime(df['time'],format ='%Y/%m/%d')
df.groupby('time')['rating'].count().plot(figsize = (12,4))
```

```python
df['month'] = df['time'].values.astype('datetime64[M]')
df.groupby('month')['rating'].count().plot(figsize=(12,4))
plt.xlabel('月份')
plt.ylabel('销售数量')
plt.title('16-19每年月销量分析')
```

```python
grouped_count_author = df.groupby('author')['frequency'].count().reset_index()
grouped_sum_amount = df.groupby('author')['amount'].sum().reset_index()
user_purchase_retention = pd.merge(left = grouped_count_author,
                                  right = grouped_sum_amount,
                                   on = 'author',
                                   how = 'inner'
                                  )
user_purchase_retention.plot.scatter(x = 'frequency',y = 'amount',figsize=(12,4))
plt.title('用户购买次数和消费金额关系图')
```

```python
df.groupby('author')['frequency'].count().plot.hist(bins= 50)
plt.xlim(1,17)
plt.xlabel('购买数量')
plt.ylabel('人数')
plt.title('用户购买门票数量直方图')
```

```python
df_frequency_2 = df.groupby('author').count().reset_index()
df_frequency_2.head()
df_frequency_2[df_frequency_2['frequency']>=2].groupby('author')['frequency'].sum().plot.hist(bins = 50)
plt.xlabel('购买数量')
plt.ylabel('人数')
plt.title('用户购买门2次及以上票数量直方图')
```

```python
df_frequency_2[df_frequency_2['frequency']>=2].groupby('frequency')['author'].count()
```

```python
df_frequency_gte_1 = df.groupby('author').count().reset_index()
#>=1
df_frequency_gte_1=df_frequency_gte_1[df_frequency_gte_1['frequency']>=1]
#<=5
values = list(df_frequency_gte_1[df_frequency_gte_1['frequency']<=5].groupby('frequency')['frequency'].count())
labels = ['1','2','3','4','5']
plt.pie(values,labels = labels,autopct = '%1.1f%%')
plt.title('购买次数1~5此之间用户占比')
plt.legend()
```

```python
df_frequency_gte_2 = df.groupby('author').count().reset_index()
df_frequency_gte_2=df_frequency_gte_2[df_frequency_gte_2['frequency']>=2]
values = list(df_frequency_gte_2[df_frequency_gte_2['frequency']<=5].groupby('frequency')['frequency'].count())
labels = ['2','3','4','5']
plt.pie(values,labels = labels,autopct = '%1.1f%%')
plt.title('购买次数2~5此之间用户占比')
plt.legend()
```

```python
pivot_count = df.pivot_table(index = 'author',
                      columns = 'month',
                       values = 'frequency',
                       aggfunc = 'count'
                      ).fillna(0)

pivot_count = pivot_count.applymap(lambda x: 1 if x > 1 else   np.NaN if x ==0 else 0)

(pivot_count.sum()/pivot_count.count()).plot()
plt.xlabel('Month')
plt.ylabel('Percentage')
plt.title('16~19 Rebuy Percentage')
```

```python
pivot_count.sum().plot()
plt.xlabel('Month')
plt.ylabel('Number')
plt.title('16~19 Rebuy Number')
```

```python
pivot_purchase = df.pivot_table(index = 'author',
              columns = 'month',
               values = 'frequency',
               aggfunc = 'count'
              ).fillna(0)
pivot_purchase.head()

def purchase_return(data):
    status = []
    for i in range(30):
        if data[i] >=1:
            if data[i+1] == 1:
                status.append(1)
            else:
                status.append(0)
        else:
            status.append(np.NaN)
        
    status.append(np.NaN)
    return pd.Series(status,pivot_purchase.columns)


pivot_purchase_return  = pivot_purchase.apply(purchase_return,axis =1)

(pivot_purchase_return.sum()/pivot_purchase_return.count()).plot()
plt.title('16~19每月回购率')
plt.xlabel('Month')
plt.ylabel('回购率Percentage')
```

```python
pivot_purchase_return.sum().plot()
plt.title('16~19每月回购人数')
plt.xlabel('Month')
plt.ylabel('回购人数')
```
